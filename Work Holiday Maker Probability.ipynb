# Analysis of Visa Data

## Loading the First Dataset

```python

import pandas as pd

# Load the first dataset
file_path = 'visas_dataF.csv'
visas_data = pd.read_csv(file_path)

# Display basic information about the dataset 
visas_data.info(), visas_data.head()
```

## Calculating Probabilities for the First Dataset

```python

# Creating a new DataFrame to store the probabilities
visas_probabilities = visas_data.copy()

# Calculate the probabilities correctly
for column in visas_data.columns:
    if column not in ['Year', 'Total']:
        visas_probabilities[column] = visas_data[column] / visas_data['Total']

# Display the first few rows of the corrected DataFrame containing the probabilities
visas_probabilities.head()
```

## Loading the Second Dataset

```python

# Load the second dataset
file_path_new = 'visas_data1F.csv'
visas_data_new = pd.read_csv(file_path_new)

# Display basic information about the new dataset and the first few rows
visas_data_new.info(), visas_data_new.head()
```

## Calculating Probabilities for 'Working Holiday Maker' Visa

```python

# Extract the Visitors.2 and Working Holiday Maker.2 columns to get P(V) and P(W) respectively.
P_V = visas_data_new['Visitors.2'] / 100  # Dividing by 100 to convert percentage to probability
P_W = visas_data_new['Working Holiday Maker.2'] / 100  # Dividing by 100 to convert percentage to probability

# Create a new DataFrame to store the probabilities of a person from a certain country applying for a Working Holiday Maker visa.
working_holiday_maker_probabilities = visas_data[['Year']].copy()

# Calculate P(W | C) for each country in each year.
for column in visas_data.columns:
    if column not in ['Year', 'Total']:
        P_C = visas_probabilities[column]  # Extract P(C) from the visas_probabilities DataFrame
        P_W_C = P_C * (P_W / P_V)  # Calculate P(W | C) using the provided formula.
        working_holiday_maker_probabilities[column] = P_W_C  # Store the calculated probabilities in the new DataFrame.

# Display the first few rows of the new DataFrame containing the calculated probabilities for Working Holiday Maker visa.
working_holiday_maker_probabilities.head()
```



- \( P(C) \) as the probability of a person from a certain country (from the first dataset).
- \( P(V) \) as the percentage of people who applied for a "Visitor" visa (from the `Visitors.2` column in the second dataset).
- \( P(W) \) as the percentage of people who applied for a "Working Holiday Maker" visa (from the `Working Holiday Maker.2` column in the second dataset).
- \( P(W | C) \) as the probability of a person from a certain country applying for a "Working Holiday Maker" visa.

Given that we know the probability of a person from a certain country applying for a "Visitor" visa, we can use the proportions of "Visitor" and "Working Holiday Maker" visas to find the probability of a person from a certain country applying for a "Working Holiday Maker" visa.

The relationship between these probabilities can be expressed as follows:

$  P(W | C) = P(C) \times \left( \frac{P(W)}{P(V)} \right) $
